```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
## Creating Line Charts

# Line chart using the MATLAB Method

# Set the x-axis to a list of strings ("") for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Line chart using the Object-Oriented Interface

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Same as above - different coding, same results
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# To create a quick graph using the object-oriented interface method, use these two lines:
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
## Using plt.show()

# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = "Boston")

# Create lables for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot and add color, width, and marker to declare parameters for lines
plt.plot(x_axis, y_axis, marker= "D", color="green", linewidth=2, label="Boston")

# Create lables for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
## Creating Bar Charts ##

# Creating Bar Charts using MATLAB Method:

# Create the plot.
plt.bar(x_axis,y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
## Create a Horizontal Bar Chart
plt.barh(x_axis,y_axis)
```

```python
# To run data on opposite axes:
plt.barh(y_axis,x_axis)
```

```python
## Create a Horizontal Bar Chart with inverted axes. gca = "get current axes"
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
# Create a Horizontal Bar Chart with coloring axes magenta and add legend
plt.barh(x_axis,y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()

# Add legend
plt.legend()

# Add title
plt.title("PyBer Fare by Month")

# Add axes labels
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
## Create Bar Charts using Object-Oriented Approach:

# Create a Vertical Bar with ax.subplots()
fig, ax =plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
## Create a Horizontal Bar Chart

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() and opposite axes.
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Create a Horizontal Bar Chart with coloring axes magenta and add legend
plt.barh(x_axis,y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()

# Add legend
plt.legend()

# Add title
plt.title("PyBer Fare by Month")

# Add axes labels
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
## Creating Scatter Plots

# Create scatter plot using MATLAB method.
plt.plot(x_axis,y_axis, "o")
```

```python
# OR/Add title, axes labels, change color, legend, etc.
plt.scatter(y_axis, x_axis, c="red", label="Chicago")

# Name axes
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Add title
plt.title("PyBer Fare By Month")

# Add limit to x-axis
plt.xlim(0, 45)

# Invert y-axis data
plt.gca().invert_yaxis()

# Add legend
plt.legend()
```

```python
## Create Bubble Chart

# Make markers larger.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

# Create the plot.
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using list comprehension to replace for and wile loops

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
## Creating a Scatter Plot using the Object-Oriented Interface

# Create the plot.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
## Using "s" parameter.

# Make markers larger.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*5)

# Create the plot.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=y_axis_larger, c="skyblue", label="Boston", alpha=0.2, linewidth=2, edgecolors="black")

# Create title
plt.title("PyBer Fare By Month")

# Create x and y names
plt.xlabel("Fares($)")
plt.ylabel("Date")

# Create legend
plt.legend()

# Add limit to x-axis.
plt.xlim(0, 45)


# Invert y-axis data.
plt.gca().invert_yaxis()
```
